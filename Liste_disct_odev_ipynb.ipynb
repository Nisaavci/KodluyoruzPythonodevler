### Ödev 2

Verilen listede bulunan string veri tipindeki öğeleri yeni_liste isimli listeye eklenir.

liste = ["Python",True,9,"3",8.4,"Hi-Kod","False",4.7]

```python
liste = ["Python", True, 9, "3", 8.4, "Hi-Kod", "False", 4.7]

yeni_liste = []

for eleman in liste:
    if type(eleman) == str:
        yeni_liste.append(eleman)

print("String öğeler:", yeni_liste)
```

### Ödev 3

Enumerate methodunu araştırın ve aşağıdaki örneği enumerate methodu ile yapın.

```python
for index in range(len(meyveler)):

    print("{}. indexte bulunan meyve: {}".format(index,meyveler[index]))
    #- enumerate() fonksiyonu, bir listeyi hem sıralı hem de indeksli şekilde dolaşmamızı sağlar.
```

```python
meyveler = ["Elma", "Muz", "Kiraz", "Portakal"]

for index, meyve in enumerate(meyveler):
    print("{}. indexte bulunan meyve: {}".format(index, meyve))
```

## Sozluk

1. Bir sözlük oluşturulur ve bu sözlükte öğrencilerin isimleri ve Matematik, Fizik, Kimya notları tutulur. Kullanıcıdan isim ve ders ismi(Matematik, Fizik, Kimya) istenir ve bu bilgilere göre çıktı verilir.

```python
# Öğrenci notlarını tutan sözlük
notlar = {
    "Nisa": {"Matematik": 90, "Fizik": 85, "Kimya": 88},
    "Ahmet": {"Matematik": 75, "Fizik": 80, "Kimya": 70},
    "Zeynep": {"Matematik": 95, "Fizik": 92, "Kimya": 89}
}

# Kullanıcıdan bilgi alma
isim = input("Öğrenci ismini girin: ")
ders = input("Ders ismini girin (Matematik, Fizik, Kimya): ")

# Bilgi kontrolü ve çıktı
if isim in notlar:
    if ders in notlar[isim]:
        print(f"{isim} adlı öğrencinin {ders} notu: {notlar[isim][ders]}")
    else:
        print("Girilen ders bilgisi geçersiz.")
else:
    print("Girilen öğrenci bulunamadı.")
```

2. Sözlük üzerinde değerleri değiştirme, yeni değer ekleme, kullanıcıya ulaşmak istediği bilgileri sorma gibi uygulamalar yapın.

```python
# Başlangıç sözlüğü
notlar = {
    "Nisa": {"Matematik": 90, "Fizik": 85, "Kimya": 88},
    "Ahmet": {"Matematik": 75, "Fizik": 80, "Kimya": 70},
    "Zeynep": {"Matematik": 95, "Fizik": 92, "Kimya": 89}
}

#  1. Değer değiştirme örneği
notlar["Ahmet"]["Fizik"] = 88  # Ahmet'in fizik notu güncellendi

#  2. Yeni öğrenci ekleme
notlar["Mehmet"] = {"Matematik": 78, "Fizik": 82, "Kimya": 80}

#  3. Var olan öğrenciye yeni ders ekleme
notlar["Nisa"]["Biyoloji"] = 91

#  4. Kullanıcıdan bilgi alarak sorgulama
isim = input("Öğrenci ismini girin: ")
ders = input("Ders ismini girin (Matematik, Fizik, Kimya, Biyoloji): ")

if isim in notlar:
    if ders in notlar[isim]:
        print(f"{isim} adlı öğrencinin {ders} notu: {notlar[isim][ders]}")
    else:
        print(f"{isim} adlı öğrencinin {ders} dersi bulunamadı.")
else:
    print("Girilen öğrenci bulunamadı.")
```
