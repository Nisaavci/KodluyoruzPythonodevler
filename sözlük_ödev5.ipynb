Sözlük:

Bir sözlük oluşturulur ve bu sözlükte öğrencilerin isimleri ve Matematik, Fizik, Kimya notları tutulur. Kullanıcıdan isim ve ders ismi(Matematik, Fizik, Kimya) istenir ve bu bilgilere göre çıktı verilir.

Sözlük üzerinde değerleri değiştirme, yeni değer ekleme, kullanıcıya ulaşmak istediği bilgileri sorma gibi uygulamalar yapın

````python
sozluk = {"Nisa" :{"fizik":80,"matematik":90,"kimya":95},
       "Ahmet": {"fizik":70,"matematik":80,"kimya":85},
       "Mehmet":{"fizik":90,"matematik":90,"kimya":95}
       }

isim=input(str("isim giriniz:")).capitalize()
ders=input(str("ders giriniz:")).lower() # Ders ismini küçük harfe çevirelim

if isim in sozluk: # İsim sözlükte var mı kontrol et
    if ders in sozluk[isim]: # Ders sözlükte var mı kontrol et
        print("notunuz {}".format(sozluk[isim][ders]))
    else:
        print(f"Üzgünüm, {isim} isimli öğrencinin '{ders}' diye bir ders notu bulunmamaktadır. Lütfen 'fizik', 'matematik' veya 'kimya' derslerinden birini girin.")
else:
    print(f"Üzgünüm, '{isim}' isimli bir öğrenci bulunmamaktadır. Lütfen doğru bir isim giriniz.")
````

Sözlük üzerinde değerleri değiştirme, yeni değer ekleme, kullanıcıya ulaşmak istediği bilgileri sorma gibi uygulamalar yapın.
